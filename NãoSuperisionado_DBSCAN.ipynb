# **Importação das Bibliotecas Necessárias**


```python
# 1. Importação das bibliotecas necessárias
import numpy as np # operações numéricas.
import matplotlib.pyplot as plt # visualizações gráficas.
from sklearn.cluster import DBSCAN # algoritmo de clustering baseado em densidade.
from sklearn.datasets import load_iris # dataset clássico de classificação.
from sklearn.metrics import adjusted_mutual_info_score, normalized_mutual_info_score # métricas de avaliação de agrupamento.
```

# **Aquisição de Dados**
300 pontos com 4 centros

```python
# 2. Carregamento do dataset Iris
iris = load_iris()
X = iris.data                # Dados de entrada (características)
labels_true = iris.target   # Rótulos reais das classes
```

# **Visualização dos Dados**

```python
# 3. Visualização inicial dos dados sem qualquer agrupamento aplicado
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], s=50, alpha=0.7)
plt.title('Dados originais sem agrupamento')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# **Aplicação do Algoritmo de Agrupamento DBSCAN**


```python
# 4. Aplicação do algoritmo DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
# eps=0.5: raio máximo para considerar um ponto vizinho.
# min_samples=5: número mínimo de pontos para formar um cluster denso.

dbscan.fit(X)
labels_pred = dbscan.labels_ #  armazena os rótulos dos clusters
```

# **Visualização dos Clusters Identificados**


```python
# 5. Visualização dos clusters identificados pelo DBSCAN
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels_pred, s=50, cmap='viridis', alpha=0.7)
plt.title('Resultados do DBSCAN Clustering')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# **Avaliação do Agrupamento**






```python
# 6. Avaliação do desempenho do agrupamento
ami = adjusted_mutual_info_score(labels_true, labels_pred) # avalia a informação compartilhada entre dois agrupamentos, ajustada para o acaso.
nmi = normalized_mutual_info_score(labels_true, labels_pred) # métrica similar, mas normalizada entre 0 e 1.

# 7. Impressão das métricas de avaliação
print(f"Adjusted Mutual Information (AMI): {ami:.4f}")
print(f"Normalized Mutual Information (NMI): {nmi:.4f}")
```
